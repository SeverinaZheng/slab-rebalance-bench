```python
import pandas as pd
df = pd.read_csv('efficiency_result_processed.csv')
```

```python
disabled_miss_ratios = df[df['rebalance_strategy'] == 'disabled'].set_index(['trace_name', 'wsr', 'allocator'])['miss_ratio']
def get_disabled_miss_ratio(row):
    key = (row['trace_name'], row['wsr'], row['allocator'])
    return disabled_miss_ratios.get(key, None)

df['disabled_miss_ratio'] = df.apply(get_disabled_miss_ratio, axis=1)
df['miss_ratio_reduction_from_disabled_ratio'] = df['miss_ratio_reduction_from_disabled'] / df['disabled_miss_ratio']
```
