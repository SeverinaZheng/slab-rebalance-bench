```python
# Filter for category == 'twitter-kv' and wsr == 0.1
print("Step 1: Initial filtering")
filtered_df = df[(df['category'] == 'twitter-kv') & (df['wsr'] == 0.1)]
print(f"Data after filtering for category='twitter-kv' and wsr=0.1: {len(filtered_df)} rows")

# Find all existing allocator-rebalance strategy combinations
print("\nStep 2: Finding existing combinations")
existing_combinations = set(filtered_df[['allocator', 'rebalance_strategy']].drop_duplicates().apply(tuple, axis=1))
print(f"Total existing allocator-rebalance strategy combinations: {len(existing_combinations)}")
print("Existing combinations:")
for combo in sorted(existing_combinations):
    print(f"  {combo[0]} + {combo[1]}")

# For each trace, check which combinations it has
print("\nStep 3: Checking completeness of each trace")
trace_combinations = {}
incomplete_traces = []

for trace_name in filtered_df['trace_name'].unique():
    trace_data = filtered_df[filtered_df['trace_name'] == trace_name]
    trace_combos = set(trace_data[['allocator', 'rebalance_strategy']].drop_duplicates().apply(tuple, axis=1))
    trace_combinations[trace_name] = trace_combos
    
    missing_combos = existing_combinations - trace_combos
    if missing_combos:
        incomplete_traces.append(trace_name)
        print(f"  {trace_name}: Missing {len(missing_combos)} combinations")
        for missing in sorted(missing_combos):
            print(f"    - {missing[0]} + {missing[1]}")
    else:
        print(f"  {trace_name}: Complete (has all {len(existing_combinations)} combinations)")

# Keep only traces that have all combinations
complete_traces = [trace for trace in trace_combinations.keys() 
                  if trace_combinations[trace] == existing_combinations]

print(f"\nStep 4: Filtering to complete traces only")
print(f"Complete traces: {len(complete_traces)} out of {len(trace_combinations)} total traces")
print("Complete traces:", complete_traces)

# Filter to only complete traces
final_df = filtered_df[filtered_df['trace_name'].isin(complete_traces)]
print(f"Final dataset: {len(final_df)} rows")

# Group by allocator and rebalance strategy, compute statistics
print("\nStep 5: Computing statistics")
if len(final_df) > 0:
    stats = final_df.groupby(['allocator', 'rebalance_strategy'])['miss_ratio_reduction_from_lru_disabled'].agg([
        'count',  # number of rows
        'min',    # minimum
        'median',
        'mean',   # mean
        'max',    # maximum
        lambda x: x.quantile(0.9)  # 90th percentile
    ]).round(6)
    
    # Rename the lambda column
    stats.columns = ['count', 'min', 'median', 'mean', 'max', 'p90']
    
    print("Statistics for miss_ratio_reduction_from_lru_disabled:")
    print("=" * 80)
    display(stats)
else:
    print("No complete traces found - cannot compute statistics")
```

```python
# Filter for meta_memcache_2024_kv trace
print("Analyzing meta_memcache_2024_kv trace")
meta_df = df[df['trace_name'] == 'meta_memcache_2024_kv']
print(f"Total rows for meta_memcache_2024_kv: {len(meta_df)}")

# Filter for tail-age and marginal-hits-tuned strategies
strategies_to_compare = ['tail-age', 'marginal-hits-tuned']
comparison_df = meta_df[meta_df['rebalance_strategy'].isin(strategies_to_compare)]

print(f"\nData available for comparison strategies: {len(comparison_df)} rows")
print(f"Available strategies: {sorted(comparison_df['rebalance_strategy'].unique())}")
print(f"Available allocators: {sorted(comparison_df['allocator'].unique())}")
print(f"Available WSR values: {sorted(comparison_df['wsr'].unique())}")

# Pivot to get miss ratios side by side for comparison
pivot_data = comparison_df.pivot_table(
    index=['allocator', 'wsr'], 
    columns='rebalance_strategy', 
    values='miss_ratio', 
    aggfunc='first'  # Since there's only one record per group
)

print("\nMiss ratios by allocator and WSR:")
print("=" * 50)
display(pivot_data.round(6))

# Calculate relative increase from marginal-hits-tuned to tail-age
if 'tail-age' in pivot_data.columns and 'marginal-hits-tuned' in pivot_data.columns:
    comparison_results = pivot_data.copy()
    comparison_results['absolute_increase'] = comparison_results['tail-age'] - comparison_results['marginal-hits-tuned']
    comparison_results['relative_increase_pct'] = ((comparison_results['tail-age'] - comparison_results['marginal-hits-tuned']) / comparison_results['marginal-hits-tuned']) * 100
    
    print("\nComparison: Tail-Age vs Marginal-Hits-Tuned")
    print("=" * 60)
    display(comparison_results.round(6))
    
    print("\nSummary of miss ratio increases (Tail-Age relative to Marginal-Hits-Tuned):")
    print("=" * 75)
    for (allocator, wsr), row in comparison_results.iterrows():
        if pd.notna(row['absolute_increase']) and pd.notna(row['relative_increase_pct']):
            tail_age_mr = row['tail-age']
            marginal_hits_tuned_mr = row['marginal-hits-tuned']
            abs_increase = row['absolute_increase']
            rel_increase = row['relative_increase_pct']
            print(f"{allocator} (WSR={wsr}):")
            print(f"  Marginal-Hits-Tuned: {marginal_hits_tuned_mr:.6f}")
            print(f"  Tail-Age: {tail_age_mr:.6f}")
            print(f"  Increase: +{abs_increase:.6f} (+{rel_increase:.2f}%)")
            print()
        else:
            print(f"{allocator} (WSR={wsr}): Missing data for comparison")
else:
    print("\nCannot perform comparison - missing strategy data")
    print(f"Available strategies: {list(pivot_data.columns)}")
```

```python
# Filter for category == 'twitter-kv', wsr == 0.01, and the three strategies
print("Step 1: Filtering data")
filtered_df = df[
    (df['category'] == 'twitter-kv') & 
    (df['wsr'] == 0.01) & 
    (df['rebalance_strategy'].isin(['tail-age', 'marginal-hits-tuned', 'disabled']))
]
print(f"Filtered data: {len(filtered_df)} rows")
print(f"Available strategies: {sorted(filtered_df['rebalance_strategy'].unique())}")
print(f"Available allocators: {sorted(filtered_df['allocator'].unique())}")
print(f"Unique traces: {len(filtered_df['trace_name'].unique())}")

# Check completeness - each (trace_name, allocator) group should have 3 strategies
print("\nStep 2: Checking data completeness")
completeness_check = filtered_df.groupby(['trace_name', 'allocator'])['rebalance_strategy'].count()
incomplete_groups = completeness_check[completeness_check != 3]

if len(incomplete_groups) > 0:
    print(f"Warning: {len(incomplete_groups)} groups don't have all 3 strategies:")
    for (trace, allocator), count in incomplete_groups.items():
        strategies = filtered_df[(filtered_df['trace_name'] == trace) & 
                                (filtered_df['allocator'] == allocator)]['rebalance_strategy'].tolist()
        print(f"  {trace} + {allocator}: only has {strategies} ({count} strategies)")
else:
    print("All groups have complete data (3 strategies each)")

# Pivot to get miss ratios side by side for comparison
print("\nStep 3: Reshaping data for comparison")
pivot_data = filtered_df.pivot_table(
    index=['trace_name', 'allocator'], 
    columns='rebalance_strategy', 
    values='miss_ratio', 
    aggfunc='first'
)

print(f"Pivot table shape: {pivot_data.shape}")
print("Sample of pivot data:")
display(pivot_data.head().round(6))

# Calculate deltas
print("\nStep 4: Calculating miss ratio deltas")
if all(strategy in pivot_data.columns for strategy in ['tail-age', 'marginal-hits-tuned', 'disabled']):
    # Calculate deltas
    pivot_data['delta_tail_age_vs_marginal_hits_tuned'] = pivot_data['tail-age'] - pivot_data['marginal-hits-tuned']
    pivot_data['relative_delta_tail_age_vs_marginal_hits_tuned'] = ((pivot_data['tail-age'] - pivot_data['marginal-hits-tuned']) / pivot_data['marginal-hits-tuned']) * 100
    pivot_data['delta_tail_age_vs_disabled'] = pivot_data['tail-age'] - pivot_data['disabled']
    
    # Remove rows with any NaN values for clean analysis
    clean_data = pivot_data.dropna()
    print(f"Clean data (no NaN): {len(clean_data)} rows out of {len(pivot_data)} total")
    
    if len(clean_data) > 0:
        # Analysis 1: Average delta (tail-age - marginal-hits-tuned)
        avg_delta_vs_tuned = clean_data['delta_tail_age_vs_marginal_hits_tuned'].mean()
        avg_relative_delta_vs_tuned = clean_data['relative_delta_tail_age_vs_marginal_hits_tuned'].mean()
        print(f"\nAnalysis 1: Tail-Age vs Marginal-Hits-Tuned")
        print("=" * 50)
        print(f"Average miss ratio delta (tail-age - marginal-hits-tuned): {avg_delta_vs_tuned:.6f}")
        print(f"Average relative delta (tail-age vs marginal-hits-tuned): {avg_relative_delta_vs_tuned:.2f}%")
        print(f"Standard deviation: {clean_data['delta_tail_age_vs_marginal_hits_tuned'].std():.6f}")
        print(f"Min delta: {clean_data['delta_tail_age_vs_marginal_hits_tuned'].min():.6f}")
        print(f"Max delta: {clean_data['delta_tail_age_vs_marginal_hits_tuned'].max():.6f}")
        
        # Analysis 2: Percent of time tail-age vs disabled is positive
        positive_vs_disabled = (clean_data['delta_tail_age_vs_disabled'] > 0).sum()
        total_comparisons = len(clean_data)
        percent_positive_vs_disabled = (positive_vs_disabled / total_comparisons) * 100
        
        print(f"\nAnalysis 2: Tail-Age vs Disabled")
        print("=" * 40)
        print(f"Cases where tail-age has HIGHER miss ratio than disabled: {positive_vs_disabled}/{total_comparisons}")
        print(f"Percentage of time tail-age performs WORSE than disabled: {percent_positive_vs_disabled:.1f}%")
        print(f"Average delta (tail-age - disabled): {clean_data['delta_tail_age_vs_disabled'].mean():.6f}")
        
        # Detailed breakdown by allocator
        print(f"\nDetailed breakdown by allocator:")
        print("=" * 60)
        for allocator in clean_data.index.get_level_values('allocator').unique():
            allocator_data = clean_data[clean_data.index.get_level_values('allocator') == allocator]
            
            avg_vs_tuned = allocator_data['delta_tail_age_vs_marginal_hits_tuned'].mean()
            avg_relative_vs_tuned = allocator_data['relative_delta_tail_age_vs_marginal_hits_tuned'].mean()
            positive_vs_disabled_alloc = (allocator_data['delta_tail_age_vs_disabled'] > 0).sum()
            total_alloc = len(allocator_data)
            percent_pos_alloc = (positive_vs_disabled_alloc / total_alloc) * 100 if total_alloc > 0 else 0
            
            print(f"{allocator}:")
            print(f"  Avg delta vs marginal-hits-tuned: {avg_vs_tuned:.6f} ({avg_relative_vs_tuned:.2f}%)")
            print(f"  Worse than disabled: {positive_vs_disabled_alloc}/{total_alloc} ({percent_pos_alloc:.1f}%)")
            
    else:
        print("No complete data available for analysis")
        
else:
    missing_strategies = [s for s in ['tail-age', 'marginal-hits-tuned', 'disabled'] if s not in pivot_data.columns]
    print(f"Cannot perform analysis - missing strategies: {missing_strategies}")
```

```python
# Group by category and WSR to compare marginal-hits-tuned vs lama
print("Step 1: Filter data for marginal-hits-tuned and lama strategies")
comparison_df = df[df['rebalance_strategy'].isin(['marginal-hits-tuned', 'lama'])].copy()
print(f"Filtered data: {len(comparison_df)} rows")
print(f"Available strategies: {sorted(comparison_df['rebalance_strategy'].unique())}")
print(f"Available categories: {sorted(comparison_df['category'].unique())}")
print(f"Available WSR values: {sorted(comparison_df['wsr'].unique())}")

# Check if we have both strategies
if len(comparison_df['rebalance_strategy'].unique()) < 2:
    print("Warning: Need both 'marginal-hits-tuned' and 'lama' strategies for comparison")
    print(f"Available strategies: {list(comparison_df['rebalance_strategy'].unique())}")
else:
    # Step 2: Pivot data to get miss ratios side by side
    print("\nStep 2: Reshaping data for comparison")
    pivot_data = comparison_df.pivot_table(
        index=['category', 'wsr', 'trace_name', 'allocator'],
        columns='rebalance_strategy',
        values='miss_ratio',
        aggfunc='first'
    )
    
    print(f"Pivot table shape: {pivot_data.shape}")
    print("Sample of pivot data:")
    display(pivot_data.head().round(6))
    
    # Step 3: Calculate comparisons
    print("\nStep 3: Computing comparisons")
    if 'marginal-hits-tuned' in pivot_data.columns and 'lama' in pivot_data.columns:
        # Add comparison column - True if marginal-hits-tuned has lower miss ratio than lama
        pivot_data['marginal_hits_tuned_better'] = pivot_data['marginal-hits-tuned'] < pivot_data['lama']
        
        # Remove rows with missing data
        clean_data = pivot_data.dropna()
        print(f"Clean data (no missing values): {len(clean_data)} rows")
        
        if len(clean_data) > 0:
            # Step 4: Group by category and wsr, compute percentage
            print("\nStep 4: Computing percentage by category and WSR")
            results = []
            
            for (category, wsr), group in clean_data.groupby(['category', 'wsr']):
                total_comparisons = len(group)
                marginal_hits_tuned_wins = group['marginal_hits_tuned_better'].sum()
                percentage_better = (marginal_hits_tuned_wins / total_comparisons) * 100
                
                avg_marginal_hits_tuned = group['marginal-hits-tuned'].mean()
                avg_lama = group['lama'].mean()
                
                results.append({
                    'category': category,
                    'wsr': wsr,
                    'total_comparisons': total_comparisons,
                    'marginal_hits_tuned_wins': marginal_hits_tuned_wins,
                    'percentage_marginal_hits_tuned_better': percentage_better,
                    'avg_miss_ratio_marginal_hits_tuned': avg_marginal_hits_tuned,
                    'avg_miss_ratio_lama': avg_lama,
                    'avg_miss_ratio_difference': avg_marginal_hits_tuned - avg_lama
                })
            
            # Convert to DataFrame for better display
            results_df = pd.DataFrame(results)
            results_df = results_df.sort_values(['category', 'wsr'])
            
            print("Results by category and WSR:")
            print("=" * 80)
            display(results_df.round(6))
            
            # Step 5: Overall summary
            print("\nStep 5: Overall Summary")
            print("=" * 50)
            overall_total = clean_data.shape[0]
            overall_wins = clean_data['marginal_hits_tuned_better'].sum()
            overall_percentage = (overall_wins / overall_total) * 100
            
            print(f"Overall across all categories and WSRs:")
            print(f"  Total comparisons: {overall_total}")
            print(f"  Marginal-hits-tuned wins: {overall_wins}")
            print(f"  Percentage where marginal-hits-tuned < lama: {overall_percentage:.2f}%")
            
            # Summary by category only
            print(f"\nSummary by category:")
            print("-" * 40)
            category_summary = clean_data.groupby('category').agg({
                'marginal_hits_tuned_better': ['count', 'sum', lambda x: (x.sum() / len(x)) * 100],
                'marginal-hits-tuned': 'mean',
                'lama': 'mean'
            }).round(6)
            
            category_summary.columns = ['total_comparisons', 'wins', 'percentage_better', 
                                      'avg_marginal_hits_tuned', 'avg_lama']
            display(category_summary)
            
            # Summary by WSR only  
            print(f"\nSummary by WSR:")
            print("-" * 40)
            wsr_summary = clean_data.groupby('wsr').agg({
                'marginal_hits_tuned_better': ['count', 'sum', lambda x: (x.sum() / len(x)) * 100],
                'marginal-hits-tuned': 'mean', 
                'lama': 'mean'
            }).round(6)
            
            wsr_summary.columns = ['total_comparisons', 'wins', 'percentage_better',
                                 'avg_marginal_hits_tuned', 'avg_lama']
            display(wsr_summary)
            
        else:
            print("No complete data available for comparison")
    else:
        missing_strategies = []
        if 'marginal-hits-tuned' not in pivot_data.columns:
            missing_strategies.append('marginal-hits-tuned')
        if 'lama' not in pivot_data.columns:
            missing_strategies.append('lama')
        print(f"Cannot perform comparison - missing strategies: {missing_strategies}")
        print(f"Available strategies: {list(pivot_data.columns)}")
```
