```python
import pandas as pd
```

```python
df = pd.read_csv('result_cycles.csv')
```

```python
latency_df = df[['allocator', 'rebalanceStrategy', 'wsr',
    'rebalancerAvgRebalanceTimeMs', 'rebalancerAvgReleaseTimeMs', 'rebalancerAvgPickTimeMs', 'rebalancerNumRebalancedSlabs']]
```

```python
# Filter for LRU2Q allocator only
lru2q_df = latency_df[latency_df['allocator'] == 'LRU2Q']

# Group by wsr and rebalanceStrategy and calculate mean and std
latency_stats = lru2q_df.groupby(['wsr', 'rebalanceStrategy'])[
    ['rebalancerAvgRebalanceTimeMs', 'rebalancerAvgReleaseTimeMs', 'rebalancerAvgPickTimeMs']
].agg(['mean']).round(4)

# Flatten column names and rename
latency_stats.columns = [
    'average_loop', 
    'average_victim_release', 
    'average_pick_victim'
]

# Display the results
print("Latency Statistics for LRU2Q (Mean and Standard Deviation):")
print("=" * 60)
latency_stats
```
