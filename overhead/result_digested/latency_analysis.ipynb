```python
import pandas as pd
```

```python
df = pd.read_csv('result_cycles.csv')
```

```python
latency_df = df[['allocator', 'rebalanceStrategy', 'wsr',
    'rebalancerAvgRebalanceTimeMs', 'rebalancerAvgReleaseTimeMs', 'rebalancerAvgPickTimeMs']]
```

```python
"""
latency_df group by allocator, rebalanceStrategy, wsr mean and std of 
'rebalancerAvgRebalanceTimeMs', 'rebalancerAvgReleaseTimeMs', 'rebalancerAvgPickTimeMs'
"""
```

```python
# Group by allocator, rebalanceStrategy, wsr and calculate mean and std
latency_stats = latency_df.groupby(['wsr', 'allocator', 'rebalanceStrategy'])[
    ['rebalancerAvgRebalanceTimeMs', 'rebalancerAvgReleaseTimeMs', 'rebalancerAvgPickTimeMs']
].agg(['mean', 'std']).round(4)

# Display the results
print("Latency Statistics (Mean and Standard Deviation):")
print("=" * 60)
latency_stats
```
